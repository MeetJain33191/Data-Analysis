## Number Guessing Game

```python
import numpy as np
import random
```

```python
n=np.random.randint(1,101)  
print("----------------------Welcome to Number Guessing Game-----------------------")
print("Press 1 for Easy\nPress 2 for Medium\nPress 3 for Hard\n")
choice=int(input("Enter your choice: "))
if choice==1:
    print("---------------Easy Level-----------------")
    c=12
    while True:
        if(c>0):
            d=int(input("Enter a number: "))  
            if(d<n):
                print("Number is too low👇")
                c-=1
                print(f"{c} moves Left")
            elif(d>n):
                print("Number is too high👆")
                c-=1
                print(f"{c} moves Left")
            else:
                print("Perfect🥳")   
                break
        else:
            print('Game-over💰')
            break
            
elif choice==2:
    print("---------------Medium Level-----------------")
    c=9
    while True:
        if(c>0):
            d=int(input("Enter a number: "))  
            if(d<n):
                print("Number is too low👇")
                c-=1
                print(f"{c} moves Left")
            elif(d>n):
                print("Number is too high👆")
                c-=1
                print(f"{c} moves Left")
            else:
                print("Perfect🥳")   
                break
        else:
            print('Game-over💰')
            break
                
elif choice==3:
    print("---------------Hard Level-----------------")
    c=6
    while True:
        if(c>0):
            d=int(input("Enter a number: "))  
            if(d<n):
                print("Number is too low👇")
                c-=1
                print(f"{c} moves Left")
            elif(d>n):
                print("Number is too high👆")
                c-=1
                print(f"{c} moves Left")
            else:
                print("Perfect🥳")   
                break
        else:
            print('Game-over💰')
            break           
```
