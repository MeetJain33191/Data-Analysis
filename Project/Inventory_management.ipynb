```python
""" 
PROJECT : 


JAY BHAVANI ::: 

                        Manager 
                        Customer 

            manager - can add products 
                    - can remove products 
                    - display updated product

        
    Enter product name :  vadapav 
    Enter qty :   10
    Enter price : 40

    {
        'vadapav' : {'qty' : 10,'price' : 40}
    }

    Do you want to enter more product ? : y 

    Enter product name :  vadapav 
    Enter qty :   30
    Enter price : 40

    {
        'vadapav' : {'qty' : 40,'price' : 40}
    }
    
    Do you want to enter more product ? : y 

    Enter product name :  Bhel 
    Enter qty :   50
    Enter price : 70

   stock 
    {
        'vadapav' : {'qty' : 40,'price' : 40},
        'bhel' : {'qty' : 50,'price' : 70}
    }

customer panel view :


Vadapav         Rs.  40
Bhel            Rs.  70


Which product you want to buy ?   vadapav

Enter no. of qty you want ?    7

Total price :  Qty : 7 X Rs. 40   = Rs. 280
----------------------------------------

for manager side stock updation :

stock 
    {
        'vadapav' : {'qty' : 33,'price' : 40},
        'bhel' : {'qty' : 50,'price' : 70
        }
    }
---------------------------------------

"""
```

```python
stock = {}

def manager_panel():
    while True:
        print("\n--- Manager Panel ---")
        print("1. Add Product")
        print("2. Remove Product")
        print("3. Display Products")
        print("4. Exit Manager Panel")
        choice = input("Enter your choice: ")

        if choice == '1':
            pname = input("Enter product name: ").lower()
            qty = int(input("Enter qty: "))
            price = int(input("Enter price: "))

            if pname in stock:
                stock[pname]['qty'] += qty
            else:
                stock[pname] = {'qty': qty, 'price': price}

            print("\nUpdated Stock:")
            print(stock)

        elif choice == '2':
            pname = input("Enter product name to remove: ").lower()
            if pname in stock:
                del stock[pname]
                print(f"{pname} removed from stock.")
            else:
                print("Product not found!")

        elif choice == '3':
            print("\nCurrent Stock:")
            for i, j in stock.items():
                print(f"{i.title()} - Qty: {j['qty']}, Price: Rs. {j['price']}")

        elif choice == '4':
            break
        else:
            print("Invalid choice!")

def customer_panel():
    print("\n--- Customer Panel ---")
    if not stock:
        print("No products available!")
        return

    print("\nAvailable Products:")
    for i, j in stock.items():
        print(f"{i.title()} - Rs. {j['price']}")

    pname = input("\nWhich product do you want to buy? ").lower()
    if pname not in stock:
        print("Product not available!")
        return

    qty = int(input(f"Enter quantity of {pname.title()} you want: "))
    if qty > stock[pname]['qty']:
        print(f"Only {stock[pname]['qty']} available. Try again.")
        return

    total = qty * stock[pname]['price']
    print(f"\nTotal price: Qty {qty} x Rs. {stock[pname]['price']} = Rs. {total}")
    print("----------------------------------------")

    stock[pname]['qty'] -= qty
    print("\nStock updated after purchase:")
    print(stock)

def main():
    while True:
        print("\nWelcome to JAY BHAVANI")
        print("1. Manager")
        print("2. Customer")
        print("3. Exit")
        role = input("Select your role: ")

        if role == '1':
            manager_panel()
        elif role == '2':
            customer_panel()
        elif role == '3':
            print("Thank you for visiting!")
            break
        else:
            print("Invalid option!")
main()
```
