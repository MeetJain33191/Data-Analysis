## 51) How Do You Traverse Through a Dictionary Object in Python? 


```python
a={'name':'Meet Jain','age':21,'city':'Ahmedabad'}
for i in a.keys():                             # traverse through keys
    print(i)
print()
for i in a.values():                             # traverse through values
    print(i)
print()
for i in a.items():                             # traverse through items(keys + values)
    print(i)
```

## 61) Write a Python function to calculate the factorial of a number (a nonnegative integer) 


```python
n=int(input("Enter a number: "))

def fact(n):
    m=1
    if(n>0):
        for i in range(1,n+1):
            m=m*i
        print(f"Factorial of {n} is {m}")
    else:
        print("Enter Positive Integer")
        
fact(n)
```

## 63) Write a Python function to check whether a number is perfect or not. 


```python
def perfect_number(num):
    sum_of_divisors = 0
    
    if num < 1:
        return False
    
    for i in range(1, num):
        if num % i == 0:
            sum_of_divisors += i 
    return sum_of_divisors == num

n = int(input("Enter number: "))
if perfect_number(n):
    print(n, "is a Perfect Number")
else:
    print(n, "is NOT a Perfect Number")
```

## 74) Write a Python program to read first n lines of a file. 


```python
filename = "intro.txt"
n = 2

def read_first_n_lines(filename, n):
    with open(filename, 'r') as file:
        for i in range(n):
            line = file.readline()
            if not line:
                break 
            print(line.strip())
read_first_n_lines(filename, n)
```

## 75) Write a Python program to read last n lines of a file. 


```python
filename = "intro.txt"
n = 2

def read_first_n_lines(filename, n):
    with open(filename, 'r') as file:
        lines = file.readlines()  
        last_lines = lines[-n:]   
        for line in last_lines:
            print(line.strip())
read_first_n_lines(filename, n)
```

## 79) Write a Python program to count the number of lines in a text file. 


```python
def count_lines(filename):
    with open(filename, 'r') as file:
        line_count = sum(1 for line in file)
    return line_count

print("Number of lines:", count_lines("intro.txt"))
```

## 90) Write python program that user to enter only odd numbers, else will raise an exception. 

```python
try:
    num = int(input("Enter an odd number: "))
    if num % 2 == 0:
        raise ValueError("You entered an even number! Only odd numbers are allowed.")
    print("You entered an odd number:", num)

except ValueError as e:
    print("Error:", e)
```
